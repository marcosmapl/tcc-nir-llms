```python
with open('ml100k-sample264-ft-falcon-7b-instruct.pkl', 'rb') as f:
    data = pickle.load(f)

data
```

```python
metrics = []

for fold in range(4):
#     print(data[f'{fold}_fold'])
    sample = []
    for item in data[f'{fold}_fold']['results']:
        if (item['gt'].lower() in item['output'].lower()):
            sample.append(1)
        else:
            sample.append(0)
    metrics.append(sample)

metrics
```

```python
cv_hit_10 = []

for i, fold in enumerate(metrics):
    print(f'Fold: {i}, fold size {len(fold)}')
    hit_10 = sum(fold) / len(fold) 
    print(f'\tHIT@10: {hit_10}')
    cv_hit_10.append(hit_10)    

mean_hit_10 = sum(cv_hit_10) / len(cv_hit_10)
print(f'Mean Hit@10: {mean_hit_10}')
```
